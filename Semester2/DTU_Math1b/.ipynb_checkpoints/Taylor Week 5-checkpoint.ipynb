```python
from sympy import *
from sympy.abc import x,y,z,u,v,w,t
from dtumathtools import *
init_printing()
```

```python
def f1(x):
    return exp(x)
def f1_1(x):
    return 1+x
def f1_2(x):
    return 1+x+0.5*x**2
def f2(x):
    return cos(x)
def f2_1(x):
    return 1
def f2_2(x):
    return 1-0.5*x**2
def f3(x):
    return exp(sin(x))
def f3_1(x):
    return 1+x
def f3_2(x):
    return 1+x+0.5*x**2
def f4(x):
    return 1/x
def f4_1(x):
    return 2-x
def f4_2(x):
    return x**2-3*x+3
```

```python
F1 = dtuplot.plot(f1,(x,-3,3), show=False)
F1.extend(dtuplot.plot(f1_1,(x,-3,3),show=False))
F1.extend(dtuplot.plot(f1_2,(x,-3,3),show=False))
F1.show()
```

```python
F2 = dtuplot.plot(f2,(x,-3,3), show=False)
F2.extend(dtuplot.plot(f2_1,(x,-3,3),show=False))
F2.extend(dtuplot.plot(f2_2,(x,-3,3),show=False))
F2.show()
```

```python
F3 = dtuplot.plot(f3,(x,-3,3), show=False)
F3.extend(dtuplot.plot(f3_1,(x,-3,3),show=False))
F3.extend(dtuplot.plot(f3_2,(x,-3,3),show=False))
F3.show()
```

```python
F4 = dtuplot.plot(f4,(x,0.5,3), show=False)
F4.extend(dtuplot.plot(f4_1,(x,0.5,3),show=False))
F4.extend(dtuplot.plot(f4_2,(x,0.5,3),show=False))
F4.show()
```
