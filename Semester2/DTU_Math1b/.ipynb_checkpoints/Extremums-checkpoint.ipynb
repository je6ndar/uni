```python
# Variables and parameters that are involved in the functions
x1, x2 = symbols('x1 x2', real=True)
a, lambda1 = symbols('a lambda1',  positive=True)
```

```python
def f1(x1, x2, a = S(1/2)):
    return a * x1**2 + 1 * x2**2

def f2(x1, x2, lambda1 = 0.5):
    Q = 1/sqrt(2) * Matrix([[1,1],[1,-1]])
    A = Q.T * Matrix([[lambda1,0],[0,1]]) * Q
    b = Matrix([-2,4])
    x = Matrix([x1,x2])
    q = x.T * A * x + x.T * b
    return q[0] 

def f3(x1, x2):
    return (1 - x1)**2 + 100*(x2 - x1**2)**2
```

```python
Nf1 = dtutools.gradient(f1(x1,x2,0.5))
Nf2 = dtutools.gradient(f2(x1,x2))
Nf3 = dtutools.gradient(f3(x1,x2))
Nf1[0]
```

```python
sols1 = nonlinsolve([Eq(Nf1[0],0), Eq(Nf1[1],0)], x1,x2)
sols2 = nonlinsolve([Eq(Nf2[0],0), Eq(Nf2[1],0)], [x1,x2])
sols3 = nonlinsolve([Eq(Nf3[0],0), Eq(Nf3[1],0)], [x1,x2])

print(sols1, f1(sols1.args[0][0], sols1.args[0][1]), 
sols2, f3(sols2.args[0][0], sols2.args[0][1]), 
sols3, f3(sols3.args[0][0], sols3.args[0][1]))
```
