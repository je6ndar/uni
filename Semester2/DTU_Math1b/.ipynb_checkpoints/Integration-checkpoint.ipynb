```python
def riemann(f,a,b,J):
    s=0
    interval = np.linspace(a,b,J+1)
    for i in interval[:-1]:
        s += f(i)*interval[1]
    return s
riemann(f, 0, 3, 2000)
```

# Regular Parametrization

## Example of the flux through the parametrized surface

```python
v = symbols("v")
V = Matrix([y*x, -x*z, x**2+y**2])
p = Matrix([u*sin(v), -u*cos(v), u*v]) # u, v [0,1]
p_jac = p.diff(u).cross(p.diff(v))
integrate(V.subs({x: p[0], y: p[1], z: p[2]}).dot(p_jac), (u, 0, 1), (v, 0, 1)).n(5)
```

# The Coulomb Vector Field

```python
V = (x**2 + y**2 + z**2)**(-3/2) * Matrix([x, y, z])
#Parametrization of a cylinder of height 2h and diameter 2a
p = Matrix([u*cos(w), u*sin(w), v]) # u [0, a], v [-h, h], w [-pi, pi]
h = 10
a = 5
#Top boundary parametrization:
p_t = Matrix([u*cos(w), u*sin(w), h]) # u [0, a], w [-pi, pi]

#Bottom boundary parametrization:
p_b = Matrix([u*cos(w), u*sin(w), -h]) # u [0, a], w [-pi, pi]

#Curved part parametrization
p_c = Matrix([a*cos(w), a*sin(w), v]) # v [-h, h], w [-pi, pi]

cyl = dtuplot.plot3d_parametric_surface(*p_c, (v, -h, h), (w, -pi, pi), show=False)
cyl.extend(dtuplot.plot3d_parametric_surface(*p_t, (u, 0, a), (w, -pi, pi), show=False))
cyl.extend(dtuplot.plot3d_parametric_surface(*p_b, (u, 0, a), (w, -pi, pi), show=False))
cyl.show()
```

## Computation of the flux through this cylinder
To do this we need to compute the flux through each surface and sum all them up.
## !REMEMBER TO CHECK THE DIRECTION OF THE NORMAL VECTORS! THEY MUST FACE AWAY FROM THE SURFACE

```python
p_jac_t = p_t.diff(u).cross(p_t.diff(w))
p_jac_b = p_b.diff(u).cross(p_t.diff(w))
p_jac_c = p_c.diff(v).cross(p_c.diff(w))
```

```python
p_jac_c, p_jac_t, p_jac_b
```

The first and the last vectors should face in the opposite direction.

Illustration of the first vector

```python
cyl.extend(dtuplot.quiver([p_c[0].subs(w,-3*pi/4), p_c[1].subs(w,-3*pi/4), 0], p_jac_c.subs(w, -3*pi/4), show=False))
cyl.show()
```

# Flux via Divergence Theorem
\begin{equation}
  \int_{\partial B} \pmb{V} \cdot \mathrm{d} \pmb{S}
  =\int_{B}\mathrm{div} (\pmb{V}) \, \mathrm{d} X.
\end{equation}
\begin{equation}
  \mathrm{div} (\pmb{V}) = \mathrm{tr} (\pmb{J}_{\pmb{V}}).
\end{equation}

```python
V = Matrix([-8*x, 8,4*z**3])
p = Matrix([r*cos(u)*sin(w), r*sin(u)*sin(w), r*cos(w)]) # u [0, pi](normally would be [0, 2pi]) ,w [0, pi]
V_par = V.subs({x: p[0], y: p[1], z: p[2]})
div_V = V.jacobian([x,y,z]).trace()

integrand = div_V.subs({x: p[0], y: p[1], z: p[2]}) * sqrt(abs((p.jacobian([r, u, w]).transpose()*p.jacobian([r, u, w])).det()).simplify())    
a = symbols("a")
flux_div = integrate(integrand, (r, 0, a),(u, 0, pi), (w, -pi/2, pi/2))
flux_div.simplify()
```

```python
dtuplot.plot3d_parametric_surface(*p.subs(r, 4), (w, 0, pi/2), (u, -pi, pi))
```

```python
p_s = Matrix([a*cos(u)*sin(w), a*sin(u)*sin(w), a*cos(w)]) # w [-pi/2, pi/2], u [0, pi]
p_b = Matrix([r*cos(u), r*sin(u), 0])
```

```python
h_s = dtuplot.plot3d_parametric_surface(*p_s.subs(a, 4), (u, -pi, pi), (w, 0, pi/2), show=False)
h_s.extend(dtuplot.plot3d_parametric_surface(*p_b, (r, 0, 4), (u, 0, 2*pi), show=False))
#h_s.camera = {"elev": -60, "azim": -50}
h_s.show()
```

```python
p_jac_s = -p_s.diff(u).cross(p_s.diff(w))
p_jac_b = -p_b.diff(r).cross(p_b.diff(u))
integrand_s = V.subs({x: p_s[0], y: p_s[1], z: p_s[2]}).dot(p_jac_s)
integrand_b = V.subs({x: p_b[0], y: p_b[1], z: p_b[2]}).dot(p_jac_b)
```

```python
flux = integrate(integrand_s, (u, -pi, pi), (w, 0, pi/2)) + integrate(integrand_b, (r, 0, a), (u, 0, 2*pi))
```

```python
flux.simplify()
```

```python
V_1 = Matrix([0 , y, 0])
V_2 = Matrix([y, -x, 0])
V_3 = Matrix([-y, x, -1])
V_4 = Matrix([x**2, x*y, x*z])

p_s = Matrix([a*cos(u)*sin(w), a*sin(u)*sin(w), a*cos(w)]) # w [-pi/2, pi/2], u [0, pi]
p_b = Matrix([r*cos(u), r*sin(u), 0])

p_jac_s = -p_s.diff(u).cross(p_s.diff(w))
p_jac_b = -p_b.diff(r).cross(p_b.diff(u))

V = [V_1, V_2, V_3, V_4]
for i in V:
    integrand_s = i.subs({x: p_s[0], y: p_s[1], z: p_s[2]}).dot(p_jac_s)
    integrand_b = i.subs({x: p_b[0], y: p_b[1], z: p_b[2]}).dot(p_jac_b)
    flux = integrate(integrand_s, (u, -pi, pi), (w, 0, pi/2)) + integrate(integrand_b, (u, 0, 2*pi))
    display(flux)
```
