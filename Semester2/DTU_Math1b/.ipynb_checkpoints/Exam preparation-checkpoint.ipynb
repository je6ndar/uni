# Orthonormal basis
## Gram-Schmidt process 
\begin{equation}
    v_n = \frac{u_n - <u_n,v_n>v_n - <u_n,v_{n-1}>v_{n-1} - \dots - <u_n,v_1>v_1}{||u_n - <u_n,v_n>v_n - <u_n,v_{n-1}>v_{n-1} - \dots - <u_n,v_1>v_1||}
\end{equation}
## d
If $S = {v_1, v_2, \dots, v_n}$ is an orthonormal basis for an inner product space $V$, and $u$ is any vector in $V$, then 
\begin{equation}
    u = <u,v_1>v_1 + <u,v_2>v_2 + \dots + <u,v_n>v_n
\end{equation}

## Common Jacobian
### Determinant for polar parametrization 
\begin{equation}
\begin{bmatrix} rcos(\theta) \\ rsin(\theta) \end{bmatrix}
\end{equation}
\begin{gather}
    J_p = \begin{bmatrix} cos(\theta) & -rsin(\theta) \\ sin(\theta) & rcos(\theta) \end{bmatrix} \\
    det(J_p) = rcos^2(\theta) + rsin^2(\theta) = r
\end{gather}
, $det(J) = r$
### Determinant for spherical coordinates
\begin{equation}
    p = \begin{bmatrix}
    rcos(u)cos(v) \\ rcos(u)sin(v) \\ rsin(u)
    \end{bmatrix}
\end{equation}
\begin{gather}
    J_p = \begin{bmatrix}
        cos(u)cos(v) &  -rsin(u)cos(v) & -rcos(u)sin(v) \\ cos(u)sin(v) & -rsin(u)sin(v) & rcos(u)cos(v) \\ sin(u) & rcos(u) & 0 
    \end{bmatrix} , \, u \in [0, \pi] , \, v \in [0, 2\pi]
\end{gather}
Using Laplace expansion to find the determinant one should get:
\begin{gather}
    det(J_p) = -rcos(u)sin(v)(rcos^2(u)sin(v) + rsin^2(u)sin(v)) - rcos(u)cos(v)(rcos^2(u)cos(v) + rsin^2(u)cos(v)) = \\
     = -r^2sin^2(v)cos(u) - r^2cos^2(v)cos(u) = -r^2cos(u) 
\end{gather}

```python
p = Matrix([r*cos(u)*cos(w), r*sin(u)*cos(w), r*sin(w)]) # u [0, 2pi], w [0, pi]
```

# Parametrization and integration in multiple dimensions
3 different ways to do it in Example below

```python
p = Matrix([v*(0.5*u**2), u])

func = 1/(y**2 + x)

#usual python func
def f_func(x, y):
    return 1/(y**2 + x)

#lambda func
func_lam = lambda x, y : 1/(y**2 + x)

#substitution of parametrization in each case
fr_sym = func.subs(zip((x,y), p)).simplify()
fr_function = f_func(*p).simplify()
fr_lam = func_lam(*p).simplify()

#Jacobian
JacobiM = Matrix.hstack(p.diff(u), p.diff(v))
Jacobian = abs(det(JacobiM)).simplify()

integral = integrate(fr_lam * Jacobian, (v, 0, 1), (u, 3/2, 5/2))
integral
```

```python
dtuplot.plot_parametric_region(*p,(v, 0, 1),(u, 3/2, 5/2))
```

```python
#this method requires all 3 coordinates to be passed(x, y, z). 0 here stands for z
dtuplot.plot3d_parametric_surface(*p, 0,(v, 0, 1),(u, 3/2, 5/2))
```

# Parametrization of a sphere
\begin{equation}
    p = \begin{bmatrix}
    rcos(u)cos(v) \\ rcos(u)sin(v) \\ rsin(u)
    \end{bmatrix}, \, u \in [0,2\pi], \, v \in [0, \pi], \, \omega \in [0, 1].
\end{equation}

```python
#parametrization
p = Matrix([r*cos(u)*cos(v), r*cos(u)*sin(v), r*sin(u)])
J_p = r**2*cos(u)
```

```python
big_ball = dtuplot.plot3d_parametric_surface(*p.subs(r,1), (u, 0, 2*pi), (v, 0, pi), aspect = 'equal', label = 'r=1', rendering_kw = {'alpha': 0.4,}, show = false)
half_ball = dtuplot.plot3d_parametric_surface(*p.subs(r,0.5), (u, 0, 2*pi), (v, 0, pi),label ='r = 0.5', rendering_kw = {'alpha': 0.5}, show = False)
quarter_ball = dtuplot.plot3d_parametric_surface(*p.subs(r,0.25), (u, 0, 2*pi), (v, 0, pi),label ='r = 0.25', show = False)
(big_ball + half_ball + quarter_ball).show()
```

### Now the integration 
the mass density function $f(x, y, z) = 1$.
\begin{equation}
    \int_p f(\boldsymbol{x}) \, d\boldsymbol{x}  = \int_{0}^{1} \int_{0}^{\pi} \int_{0}^{2\pi} f(\mathbf{p}) |det(J_p)| \, du dv dr
\end{equation}

```python
f = lambda x, y, z: 1
integrand = (f(*p) * abs(J_p)).simplify()
mass = integrate(integrand, (u, 0, 2*pi), (v, 0, pi), (r, 0, 1))
integrand, mass
```

# Vector field along stair line
\begin{equation}
    \mathbf{V}(x_1, x_2) = (x_1x_2,x_1)
\end{equation}
The line goes from the origin to the point $(x_1, x_2)$. The idea is to say that the line is the vector $\begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ then the parametrization will be:
\begin{equation}
    \underline{r} = \begin{bmatrix} ux_1 \\ ux_2\end{bmatrix}, \, u \in [0, 1]
\end{equation}

# The anti-derivative problem
\begin{equation}
    \mathbf{V}(x_1,x_2,x_3) = \begin{bmatrix} x_2 cos(x_1x_2) \\ 
        x_3 + x_1 cos(x_1 x_2) \\
        x_2
    \end{bmatrix}
\end{equation}
\begin{equation}
    \mathbf{W}(x_1,x_2,x_3) =\frac{1}{1+x_1^2 x_2^2 + 2x_1 x_2 x_3^2 + x_3^4} \begin{bmatrix} x_2 \\ 
        x_1 \\
        2x_3
    \end{bmatrix}
\end{equation}

Lemma 7.3.2 - Jacobian matrix must be symmetrical for the grad vec field

```python
V = Matrix([x2 * cos(x1*x2), x3 + x1*cos(x1*x2), x2])
W = Matrix([x2, x1, 2*x3])/(1+x1**2 * x2**2 + 2*x1*x2*x3**2 + x3**4)
```

```python
Jac_V = V.jacobian([x1, x2, x3])
Jac_W = W.jacobian([x1, x2, x3])
display(Jac_V, Jac_W)
```

To find all antiderivatives we can compute line integral along the line.

```python
x1, x2, x3 = symbols("x1, x2, x3", real=True, nonnegative=True)
```

```python
r = u*Matrix([x1, x2, x3])
integrand = V.subs({x1: r[0], x2: r[1], x3: r[2]}).dot(r.diff(u)) 
integrate(integrand, (u, 0, 1))
```
