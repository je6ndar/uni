# Exercise 1

Consider the quadratic form $q: \mathbb{R}^3 \to \mathbb{R}$ given by
\begin{equation*}
q(\pmb{x}) = 5 x_{1}^{2} + 5 x_{2}^{2} + 8 x_{3}^{2} + 8 x_{1} x_{2} - 4 x_{1} x_{3} + 4 x_{2} x_{3} - 22 x_{1} - 32 x_{2} - 20 x_{3} + 53,
\end{equation*}
where $\pmb{x} = [x_1,x_2,x_3]^T \in \mathbb{R}^3$.

```python

xvec = Matrix([x1,x2,x3])
q = 5*x1**2 + 8*x1*x2 - 4*x1*x3 - 22*x1 + 5*x2**2 + 4*x2*x3 - 32*x2 + 8*x3**2 - 20*x3 + 53
```

## a

Compute the gradient $\nabla q(\pmb{x})$ for any $\pmb{x} \in \mathbb{R}^3$.

To find the gradient we need to compute the partial derivatives.

```python
grad = Matrix([q.diff(x1), q.diff(x2), q.diff(x3)]).transpose()
grad
```

## b

Compute the Hessian matrix $\pmb{H}_q$. *Hint: The Hessian matrix should not depend on $\pmb{x}$.*

The hessian matrix of the given function will look like this, but it is going to be 3x3 matrix as we need to add partial derivatives with respect to the third variable:
\begin{equation}
    H_f(x, y) = \begin{bmatrix} \frac{\partial^2{f}}{\partial{x^2}} & \frac{\partial^2{f}}{\partial x \partial y} \\ \frac{\partial^2{f}}{\partial y \partial x} & \frac{\partial^2{f}}{\partial{y^2}} \end{bmatrix}
\end{equation}

```python
H_q = dtutools.hessian(q)
H_q
```

## c

Find an orthonormal basis of eigenvectors for the Hessian matrix $\pmb{H}_q$.

According to the spectral theorem if matrix $H_q$ is symmetric(it is), a real orthogonal matrix $Q$ exists such that $Q^{T}AQ=D$, where $D$ is a diagonal matrix of eigenvalues.

```python
H_q.diagonalize()
```

We can see that $\lambda = 18$ has multiplicity of 2, so we need to check if the eigenvectors are orthogonal and normalize them. Also eigenvectors that belong to different eigenvalues are already orthonormal.

```python
v1 = Matrix([2, -2, 1])
v2 = Matrix([1, 1, 0])
v3 = Matrix([-1, 0, 2])
```

```python
v2.dot(v3)
```

As we now know they are not orthogonal we can apply gram-schmidt process to orthonormalize them.

```python
v3, v2 = GramSchmidt([v3,v2], orthonormal=True)
v3, v2
```

And we also need no normalize the first vector.

```python
v1 = v1/v1.norm()
```

```python
Q = Matrix.hstack(v1, v2, v3)
Q
```

So, the $Q$ is the orthonormal basis of eigenvectors for the Hessian matrix $\pmb{H}_q$.

## d

Show that $(1,2,1)$ is a stationary point of $q$. Find all stationary points of $q$.

To find stationary points of $q$ we need to equate gradient to 0.

```python
eq1 = Eq(grad[0],0)
eq2 = Eq(grad[1],0)
eq3 = Eq(grad[2],0)
```

```python
solve([eq1, eq2, eq3], x1, x2, x3)
```

We can write it as
\begin{equation}
    \begin{bmatrix} x1 \\ x2 \\ x3 \end{bmatrix} = \begin{bmatrix} 2 \\ -2 \\ 1\end{bmatrix} t + \begin{bmatrix} -1 \\ 4\\ 0 \end{bmatrix}
\end{equation} , where $t \in \mathbb{R}$

## e

State a direction from the stationary point $(1,2,1)$ in which the function $q$ neither increases nor decreases (i.e., a direction where the function is constant).

To find this direction we need find a unit vector such that the dot product of this vector and the gradient vector at the given point is $0$

```python
grad.subs({x1: 1, x2: 2, x3: 1})
```

As the gradient itself equals to 0 we can state that the point $(1, 2, 1)$ is stationary point and the function at this point is neither increasing nor decreasing in any direction.

Consider the quadratic form $q: \mathbb{R}^4 \to \mathbb{R}$ given by

q(\pmb{x}) = 2 x_{1} x_{3} + 4 x_{2} x_{4},

## a

State a *symmetric* matrix $A \in \mathbb{R}^{4 \times 4}$ such that $q(\pmb{x}) = \pmb{x}^T A \pmb{x}$ , where $\pmb{x}=[x_1,x_2,x_3,x_4]^T$.

```python
# Define the matrix A
A = Matrix([[0,0,1,0],[0,0,0,2],[1,0,0,0],[0,2,0,0]])

# Define a vector of the variables
var = Matrix([x1, x2, x3, x4])

# Display all the matrix A and the quadratic equation
display(A, var.transpose()*A*var)
```

## b

Find an orthogonal (change-of-basis) matrix $Q\in \mathbb{R}^{4 \times 4}$ that "reduces" the quadratic form $q$ in the sense that $q$, in the new coordinates, does not contain "mixed terms" of the form $x_i x_j$ (where $i \neq j$). Express $q$ in the new coordinates.

In orded to achive reduced quadratic form we need to find orthonormal basis $\beta = u_1, u_2, u_3, u_4$ for $\underline{\underline{A}}$ such that 
\begin{align*}
&(\underline{\underline{U}} * \underline{\tilde{x}})^T * \underline{\underline{A}} * (\underline{\underline{U}} * \underline{\tilde{x}}) + (\underline{\underline{U}} * \underline{\tilde{x}})^T * \underline{b} + c = 0 \\
&\underline{\tilde{x}}^T * \underline{\underline{U}}^T * \underline{\underline{A}} * \underline{\underline{U}} * \underline{\tilde{x}} + \underline{\tilde{x}}^T * \underline{\underline{U}}^T * \underline{b} + c = 0 \\
&\underline{\tilde{x}}^T * \underline{\underline{D}} * \underline{\tilde{x}} + \underline{\tilde{x}}^T * \underline{\underline{U}}^T * \underline{b} + c = 0,
\end{align*}
where $\underline{\underline{U}} = \begin{bmatrix} u_1 & u_2 & u_3 & u_4\end{bmatrix}$ and $\underline{\underline{D}}$ is a diagonal matrix of corresponding eigenvalues of $\underline{\underline{U}}$.

```python
#Find the eigenvectors and eigenvalues of A
eigens = A.eigenvects() #returns list of tuples with eigenvalue, its multiplicity and eigenvectors

eigens
```

```python
lamda1 = eigens[0][0] #first eigenvalue
lamda2 = eigens[1][0] #second eigenvalue
lamda3 = eigens[2][0] #third eigenvalue
lamda4 = eigens[3][0] #forth eigenvalue

u1 = eigens[0][2][0] #first eigenvector lamda1
u2 = eigens[1][2][0] #second eigenvector lamda2
u3 = eigens[2][2][0] #third eigenvector lamda3
u4 = eigens[3][2][0] #forth eigenvector lamda4
```

As $\underline{\underline{A}}$ is symmetric then
1) Diagonalization is possible.
2) The different $E_{\lambda}$(eigenvectors belonging to different $\lambda$) are orthogonal.

As all of 4 eigenvalues have only one corresponding eigenvector the eigenvectors are already orthogonal, and we only need to normalize them.

```python
u1 = u1/u1.norm()
u2 = u2/u2.norm()
u3 = u3/u3.norm()
u4 = u4/u4.norm()

Q = Matrix([u1.T, u2.T, u3.T, u4.T]).T
D = diag(lamda1, lamda2, lamda3, lamda4)
```

The matrix $Q$ is

```python
Q
```

The $q$ expressed in new coordinates is

```python
q_new = var.transpose()*D*var
q_new
```

## d

Show that
\begin{equation*}
    \lim_{x \to 0} f(x,x) = \frac{1}{2}.
\end{equation*}

To find the limit we can use following:

one can say that $cos(x) = P_2(x) + R_2(x) = P_2(x) + \epsilon(x)x^2$. An $\epsilon$ is a category of function that fulfill:
1) $\epsilon(0) = 0$
2) $\epsilon(x) \rightarrow 0$

So, using the result from the previous question we can write
\begin{equation}
    \lim_{x \to 0} f(x,x) = \lim_{x \to 0} \frac{x^2(1-\frac{x^2}{2} + \epsilon(x)x^2)}{x^2 + x^2} = \frac{1}{2}(1 - x^2 + \epsilon(x)x^2))
\end{equation}
As $x$ goes to $0$, we can conclude:
\begin{equation}
     \lim_{x \to 0} f(x,x) = \frac{1}{2}(1) = \frac{1}{2}
\end{equation}

## e

Determine the limit
\begin{equation*}
    \lim_{x \to 0} f(x,2x). 
\end{equation*}

Using the same logic,
\begin{equation}
    \lim_{x \to 0} f(x,2x) = \lim_{x \to 0} \frac{4x^2(1-\frac{x^2}{2} + \epsilon(x)x^2)}{x^2 + 4x^2} = \lim_{x \to 0} \frac{4x^2(1-\frac{x^2}{2} + \epsilon(x)x^2)}{5x^2} = \lim_{x \to 0} \frac{4(1-\frac{x^2}{2} + \epsilon(x)x^2)}{5} = \frac{4}{5}
\end{equation}

# Exercise 4

Consider the vector field $\pmb{V}: \mathbb{R}^3 \to \mathbb{R}^3$ given by
\begin{equation*}
    \pmb{V}(x,y,z) = (-x, x y^2, x+z)
\end{equation*}
and the curve $\mathcal{K}_1$ given by the parametrization:
\begin{equation*}
    \pmb{r}(u) = (u, u^2, u+1), \quad u \in [0,2].
\end{equation*}
Thus, $\mathcal{K}_1 = \mathrm{im}(\pmb{r})$.

```python
x, y, z, u, t = symbols("x y z u t", real=True)
V = Matrix([-x, x*y**2, x+z])
r = Matrix([u, u**2, u +1])
```

## a

Find the tangent vector $\pmb{r}'(u)$ and argue that the parameterization is regular.

```python
r_t = r.diff(u)
r_t
```

According to the Definition 3.1.4 if $\mathbf{r'}(t) \neq \mathbf{0}$ the parametrization is regular.

So, it is evident that this parametrization is regular.

## b

Calculate $\langle \pmb{V}(\pmb{r}(u)), \pmb{r}'(u) \rangle$ for any $u \in [0,2]$ and compute the line integral $\int_{\mathcal{K}_1} \pmb{V} \cdot \,\mathrm{d} \pmb{s}$.

According to the Definition 7.4.1 the dot product $\langle \pmb{V}(\pmb{r}(u)), \pmb{r}'(u) \rangle$ is the integrand itself. To compute the line integral $\int_{\mathcal{K}_1} \pmb{V} \cdot \,\mathrm{d} \pmb{s}$ we need integrate the dot product with respect to $u$.

```python
# The dot product
integrand = V.subs({x: r[0], y: r[1], z: r[2]}).dot(r_t)

# Integration
int_V = integrate(integrand, (u, 0, 2))

display(integrand, int_V.evalf())
```

## c

Find a parametrization $\pmb{p}:[0,1] \to \mathbb{R}^3$ of the straight line from $(0,0,1)$ to $(2,4,3)$. We denote the line segment by $\mathcal{K}_2=\mathrm{im} (\pmb{p})$. Compute the line integral $\int_{\mathcal{K}_2} \pmb{V} \cdot \,\mathrm{d} \pmb{s}$.

First we parametrize the line. We can do it by thinking of the line in terms of the vector $\begin{bmatrix} 2 \\ 4 \\ 3\end{bmatrix} - \begin{bmatrix} 0 \\ 0 \\ 1\end{bmatrix}$, which we can multiply by the coefficient $u$, $u \in [0, 1]$, the only step we are missing is the starting point, which can be achieved by adding $\begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$ to our vector.

\begin{equation}
    p = \begin{bmatrix} 2u \\ 4u \\ 2u + 1 \end{bmatrix}
\end{equation}

```python
p = Matrix([u*2, u*4, u*2 + 1])
p
```

We can varify it by plotting

```python
dtuplot.plot3d_parametric_line(*p,(u,0,1))
```

The line integral can be computed in the same manner:

```python
#The tangential vector
p_dif = p.diff(u)

#dot product of the vector field and the tangential vector
integrand = V.subs({x: p[0], y: p[1], z: p[2]}).dot(p_dif)

integrate(integrand, (u, 0, 1))
```
