```python
import numpy as np
from sympy import *
from dtumathtools import*
```

```python
u, v, w = symbols("u, v, w")
```

```python
r = Matrix([[0.5*u**2 - v**2],[-u*v],[w]])
u_range = (u, 0, 2)
v_range = (v, 0, 2)
w_range = (w, 0, 2)
plt = dtuplot.plot3d_parametric_surface(*r.subs(w,0), u_range, v_range, show=False)
plt.extend(dtuplot.plot3d_parametric_surface(*r.subs(w,1), u_range, v_range, show=False))
plt.extend(dtuplot.plot3d_parametric_surface(*r.subs(w,2), u_range, v_range, show=False))
plt.show()
```
