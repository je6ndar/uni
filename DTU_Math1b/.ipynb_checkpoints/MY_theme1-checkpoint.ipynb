```python
from dtumathtools import *
from sympy import *
init_printing()
```

```python
# Variables and parameters that appear in the functions
x1, x2, t = symbols('x1 x2, t', real=True)
a, lambda1 = symbols('a lambda1',  positive=True)
```

```python
def f1(x1, x2, a = 0.5):
    return a * x1**2 + 1 * x2**2

def f2(x1, x2, lambda1 = 0.5):
    Q = 1/sqrt(2) * Matrix([[1,1],[1,-1]])
    A = Q.T * Matrix([[lambda1,0],[0,1]]) * Q
    b = Matrix([-2,4])
    x = Matrix([x1,x2])
    q = x.T * A * x + x.T * b
    return q[0] 

def f3(x1, x2):
    return (1 - x1)**2 + 100*(x2 - x1**2)**2
```

```python
f1(x1,x2), f1(x1,x2,2), f1(x1,x2,a)
```

```python
f3(x1,x2).expand()
```

```python
def gradient_method(f, x_0, alpha = 0.01, tol = 0.0001, n_max = 5000):
    nf = dtutools.gradient(f, (x1,x2))
    x = x_0
    nfx = nf.subs({x1: x_0[0], x2: x_0[1]})
    n = 0
    while nfx.norm() > tol and n < n_max:
        x = x - alpha * nfx
        nfx = nf.subs({x1: x[0], x2:x[1]})
        n += 1
        #print(x)
    return x, n
gradient_method(f1(x1,x2,0.5), Matrix([1,2]), 0.01)
#gradient_method(f1(x1,x2,0.5), x_0 = Matrix([1,10]), alpha = 0.1, tol = 1e-4)
```
