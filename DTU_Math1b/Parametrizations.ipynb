### Hollow Circle (2D)

```python
from sympy import *
from dtumathtools import *

u = symbols("u")

R = 1 # radius

a,b = (0,0) # center coordinate

x = R*cos(u)+a
y = R*sin(u)+b
r_hollow_circle = Matrix([x, y])

display(r_hollow_circle)
p = dtuplot.plot_parametric(*r_hollow_circle, (u, 0, 2*pi),show=False, use_cm=False)
p.xlabel = "$x$"
p.ylabel = "$y$"
p.ax.set_aspect("equal")
```

According to Equation 7.8, the line integral of a parameterized curve $C$ is defined as follows:

$$ \int_C f(\textbf{x}) \ ds = \int_a^b f(\textbf{x}) \ ||\textbf{r}'(u) || \ du $$

, where the length of the curve is found by:

$$ length(C) = \int_C 1 \ ds = \int_a^b ||\textbf{r}'(u) || \ du $$

```python
jacobi = r_hollow_circle.diff(u).norm() # jacobi func
# J = r.jacobian((u,v))
# jacobi = sqrt(det(J.T*J)) # jacobi func for all cases
integrate(jacobi, (u, 0, 2*pi))
```

### Filled Circle

```python
from sympy import *
from dtumathtools import *
u = symbols("u")
v = symbols("R")

a,b,c = (0,0,0) # center coordinate

x = v*cos(u)+a
y = v*sin(u)+b
z = c
r_filled_circle = Matrix([x, y, z])

display(r_filled_circle)
p = dtuplot.plot3d_parametric_surface(*r_filled_circle, (u, 0, 2*pi), (v,0,1))
```

According to Equation 7.11, the surface integral of a parameterized surface $F$ by vectors $(u,v)$ in $\Gamma$ is defined as follows:

$$ \int_F f(\textbf{x}) \ ds = \int_{\Gamma} f(\textbf{r}(u,v)) \ || r_u' \  \ r_v' || \ d(u,v)$$
, where the area of the surface is found by:
$$area(F) = \int_F 1 \ dS = \int_{\Gamma} || r_u' \  \ r_v' || \ d(u,v)$$

```python
r = r_filled_circle
jacobi = r.diff(u).cross(r.diff(v)).norm() # jacobi func
# J = r.jacobian((u,v))
# jacobi = sqrt(det(J.T*J)) # jacobi func for all cases
integrate(jacobi, (u, 0, 2*pi), (v,0,1))
```

### Surface of Ellipsoid

```python
from sympy import *
from dtumathtools import *
u,v = symbols("u v")

a,b,c = (0,0,0) # center coordinate

# Sphere when R=A=B=C
A = 2
B = 2
C = 2

x = A*sin(u)*cos(v) + a
y = B*sin(u)*sin(v)+b
z = C*cos(u)+c

r_ellipsoid_surface = Matrix([x, y, z])

p = dtuplot.plot3d_parametric_surface(*r_ellipsoid_surface, (u, 0, pi), (v,0,2*pi),show=False)
p.ax.set_aspect("equal")
display(r_ellipsoid_surface)
```

According to Equation 7.11, the surface integral of a parameterized surface $F$ by vectors $(u,v)$ in $\Gamma$ is defined as follows:

$$ \int_F f(\textbf{x}) \ ds = \int_{\Gamma} f(\textbf{r}(u,v)) \ || r_u' \  \ r_v' || \ d(u,v)$$
, where the area of the surface is found by:
$$area(F) = \int_F 1 \ dS = \int_{\Gamma} || r_u' \  \ r_v' || \ d(u,v)$$

```python
r = r_ellipsoid_surface
jacobi = r.diff(u).cross(r.diff(v)).norm() # jacobi func
#J = r.jacobian((u,v))
#jacobi = sqrt(det(J.T*J)) # jacobi func for all cases
integrate(jacobi, (u, 0, pi), (v,0,2*pi)).doit().doit()
```
