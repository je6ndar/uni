```python
def funFdF(X):
    x1, x2 = X
    F = np.array([x1**2 + 2*x2 - 2, x1 + 4*x2**2-4])
    dF = np.array([[2*x1, 2], [1,8*x2]])
    return F, dF
```

```python
def Newton_Sys(FdF, X, n):
    F, dF = FdF(X)
    X_iter = []
    for i in range(n):
        H = np.linalg.solve(dF,F)
        X = X - H
        X_iter.append(X)
        F, dF = FdF(X)
    return X_iter
```

```python
X0 = np.array([1,2])
X = Newton_Sys(funFdF, X0, 8)
X[-1]
```

```python
X_last = X[-1]
err = []
for x in X[:-1]:
    norm = np.linalg.norm(X_last - x)
    err.append(norm)
```

```python
n = np.linspace(1, 7, 7)
plt.semilogy(n, err)
plt.xlabel('n')
plt.ylabel('||X* - X||')
plt.title('Absolute error vs iteration')
plt.grid(True)
```
