### Experiment




We consider the problem of discovering a model $y=f(h,g,t)$ where $y$ is the height of an object at time $t$, and where $h$ is the initial height of the object, and 
$g$ is the acceleration of gravity.

We note that only length and time dimensions are present in the arguments of the function. We thus need to find a characteristic length scale, $\lambda$,
and a characteristic time scale $\tau$.

Formally, these can be determined by solving
$\lambda=h^Ag^Bt^C$ and 
$\tau=h^Dg^Et^F$.
This leads to the linear equations $A+B=1$ and $-2B+C=0$ for $\lambda$,
and $D+E=0$ and $-2E+F=1$ for $\tau$. One solution is $A=1$, $B=0$, $C=0$, $D=0$, $E=0$ and $F=1$ which gives 
$\lambda=h$
and 
$\tau=t$.

We insert these characteristic scales into the model:

$\frac{y}{\lambda}=f\left(\frac{h}{\lambda},\frac{g}{\frac{\lambda}{\tau^2}},\frac{t}{\tau}\right)$

$\frac{y}{h}=f\left(\frac{h}{h},\frac{g}{\frac{h}{t^2}},\frac{t}{t}\right)$

$\frac{y}{h}=f\left(1,\frac{gt^2}{h},1\right)$

$\frac{y}{h}=f\left(\frac{gt^2}{h}\right)$

We end with to dimensionless variables, $\pi_1=\frac{y}{h}$ and $\pi_2=\frac{gt^2}{h}$: $\pi_1=f(\pi_2)$.




#### Generate "experimental" data

Below we generate "experimental" data with values.
We select random values for times and initial heights from uniform distributions.
With these values we compute the height $y$ with some added normally distribute noise. 

```python
import numpy as np
rng = np.random.default_rng(523572346764)
g = 9.82 # acceleration of gravity
t = rng.uniform(low=0.2,high=0.8,size=10) # random times between 0.2 s and 0.8 s
h = rng.uniform(low=2.0,high=5.0,size=np.size(t)) # random inital heights between 2.0 m and 5.0 m
y = h - 1/2 * g * t**2 + +rng.normal(loc=0.0,scale=0.05,size=np.size(t)) # compute heights with added normally distributed noise
print('t='+str(t),'h='+str(h),'y=',str(y))
```

#### Compute the dimensionless variables and plot them

We compute the values for the dimensionless variables and plot them.
We notice a linear tendency in the data, so we suggest a model $\pi_1=A+B\pi_2$. 

```python
pi_2 = g*t**2/h
pi_1 = y/h
import matplotlib.pyplot as plt
plt.figure()
plt.plot(pi_2,pi_1,'b+')
plt.xlabel('$\pi_2$ (1)')
plt.ylabel('$\pi_1$ (1)')
plt.show()
```

#### Perform regression analysis

We perform a regression analysis adn note that the coefficients, $A$ and $B$ are close to $1$ and $-\frac{1}{2}$.
We compute the standard score and it is less than one uncertainty so we cannot reject that $A=1$ and $B=-\frac{1}{2}$.

```python
from scipy.optimize import curve_fit
def f(x,A,B):
    return A+B*x
param,cov = curve_fit(f,pi_2,pi_1)
print(param)
print(np.sqrt(np.diag(cov)))
import uncertainties as unc
A = unc.ufloat(param[0],np.sqrt(np.diag(cov))[0])
B = unc.ufloat(param[1],np.sqrt(np.diag(cov))[1])
print('A = {}'.format(A))
print('B = {}'.format(B))
print(-A.std_score(1.0))
print(-B.std_score(-0.5))
```

#### Plot of data and regression model

```python
x_values = np.linspace(min(pi_2),max(pi_2),100)
plt.figure()
plt.plot(pi_2,pi_1,'b+')
plt.plot(x_values,f(x_values,*param),'r--')
plt.show()
```

#### Find final model

$\frac{y}{h}=A+B\frac{gt^2}{h}$

The linear regression analysis suggested that $A=1$ and $B=-\frac{1}{2}$. Inserting these in the model gives

$\frac{y}{h}=1-\frac{1}{2}\frac{gt^2}{h}$

and isolating $y$ yields the final model

$y=h-\frac{1}{2}gt^2$
