## How to make graphs in Python

### Plotting points and lines

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = default_rng(seed=1000) # create a random number generator
N = 11
t = np.linspace(0,N-1,N)
x = t**2 + 0.5 * rng.normal(loc=0.0,scale=1.0,size=N) # add normally distributes values
plt.figure()
plt.plot(t,x,'r+',markersize=10)
plt.show()
plt.plot(t,x,'b-',linewidth=2)
plt.show()
plt.plot(t,x,'k--')
plt.plot(t,x,'ro')
plt.show()
````

### Plotting data with uncertainties
When plotting data with uncertainties you can use *errorbar* insted of *plot*.
The standard use is as shown in the example below.
```python
plt.errorbar(x,y,errx=dx,erry=dy,fmt='.')
```
You can leave out either *errx* or *erry* if you don't have values for them.

The uncertaintines *dx* and *dy* can either be a value of the uncertainty (assumed equal for all data points) or an array of individual uncertainties.
With *fmt='.'* only data points and uncertainties are plotted. To get lines between point use Use *fmt='.-'*.

To close the uncertainties use e.g. *capsize=3*.

````python
dy = np.full(N,fill_value=0.6)
plt.errorbar(t,x,yerr=0.4,fmt='o',capsize=3)
plt.title('Identical uncertainties in $y$-values')
plt.show()
from numpy.random import default_rng
rng = default_rng(seed=156666664) # create a random number generator
dx = rng.normal(loc=0.3,scale=0.1,size=N)
plt.title('Individual uncertainties in $x$- and $y$-values')
plt.errorbar(t,x,xerr=dx,yerr=dy,fmt='o',capsize=3)
plt.show()
````

### Logarithmic plots

1. *semilogx* makes the first axis logarithmic.
1. *semilogy* makes the second axis logarithmic.
1. *loglog* makes both axes logarithmic.

````python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,10,1000)
x = np.exp(-t)
plt.figure()
plt.semilogx(t,x,'k-')
plt.show()
plt.semilogy(t,x,'k-')
plt.show()
plt.loglog(t,x,'k-')
plt.show()
````

### Plotting with labels, units, and legends


````python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,10,1000)
x = 2.0*np.cos(t)
v = -2.0*np.sin(t)
a = -2.0*np.cos(t)
ax = plt.axes()
ax.plot(t,x,'b-',label='Large oscillation')
ax.plot(t,v,'m-',label='Small oscillation')
ax.legend()
plt.xlim(0,10)
plt.ylim(-2.1,2.1)
plt.xlabel('Time $t$ (s)')
plt.ylabel('Positiontion $x,y$ (m)')
plt.show()
````

````python
plt.plot(v,a,'b-')
plt.xlabel('Velocity $v_x$ (m/s)')
plt.ylabel('Acceleration $a_x$ (m/s$^2$)')
plt.show()
````

#### Other useful elements

1. *xlim(0,10)* and *ylim(-3,3)* specifies the ranges on the axes.
1. *text()* plots a text at a certain point.
1. *grid(True)'* draws a grid on the figure. This can, e.g., help reading of values.
1. *axis('scaled')* uses the same scale on both axes, such that circles look circular and not elliptical.
1. To plot an arrow use *plt.arrow(1, 0.2, -0.2, 0.2, head_width=0.05, head_length=0.1, fc='k', ec='k')*.
1. *figsize(,)* makes the figures of the requested size.
1. *subplots* makes several plots in one figure.
1. *plt.subplots(3,1,layout='constrained')* creates three plots in one column.
1. To plot in a subplot use *plt.subplot(311)* which selects the first plot.

````python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,10,1000)
x = 2.0*np.cos(t)
v = -2.0*np.sin(t)
a = -2.0*np.cos(t)
ax = plt.axes()
ax.plot(t,x,'b-',label='Large oscillation')
ax.plot(t,v,'m-',label='Small oscillation')
ax.legend()
plt.xlim(0,10)
plt.ylim(-3,3)
plt.xlabel('Time $t$ (s)')
plt.ylabel('Positiontion $x,y$ (m)')
plt.grid(True)
plt.axis('scaled')
plt.text(2.0,1.0,r"Text: $\theta^4$")
plt.arrow(6, -1, -0.9, 0.7, head_width=0.15, head_length=0.3, fc='k', ec='k')
plt.show()
t = np.linspace(0,4*np.pi,500)
plt.subplots(3,1,layout='constrained')
plt.subplot(311)
plt.xlabel('$t$')
plt.ylabel(r'$\sin(t)$')
plt.plot(t,np.sin(t),'r-')
plt.subplot(312)
plt.xlabel('$t$')
plt.ylabel(r'$\cos(t)$')
plt.plot(t,np.cos(t),'r-')
plt.subplot(313)
plt.xlabel('$t$')
plt.ylabel(r'$\exp(-t/5)\cdot\sin(3*t)$')
plt.plot(t,np.exp(-t/5)*np.sin(3*t),'r-')
plt.plot(t,np.exp(-t/5),'b--',linewidth=1.0)
plt.plot(t,-np.exp(-t/5),'b--',linewidth=1.0)
plt.show()
````

### Putting it all together

#### Plot data points

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = default_rng(seed=15256254254) # create a random number generator
N = 10
t = np.linspace(0,1,N)
x = t**2+rng.normal(loc=0,scale=0.05,size=N)
plt.figure()
plt.plot(t,x,'bo')
plt.show()
````

#### Add uncertainties

````python
dt = rng.normal(loc=0.05,scale=0.01,size=np.size(t))
dx = rng.normal(loc=0.1,scale=0.02,size=np.size(t))
plt.errorbar(t,x,xerr=dt,yerr=dx,fmt='o',capsize=4,label='Data points with uncertainties')
plt.show()
````

#### Add regression curve

````python
t2 = np.linspace(0,1,10*N)
x2 = t2**2
plt.errorbar(t,x,xerr=dt,yerr=dx,fmt='o',capsize=4,label='Data points with uncertainties')
plt.plot(t2,x2,'k--',label='Regression curve')
plt.show()
````

#### Add labels to the axes and a grid

````python
t2 = np.linspace(0,1,10*N)
x2 = t2**2
plt.errorbar(t,x,xerr=dt,yerr=dx,fmt='o',capsize=4,label='Data points with uncertainties')
plt.plot(t2,x2,'k--',label='Regression curve')
plt.xlabel('Time $t$ (s)')
plt.ylabel('Acceleration $a$ (m/s$^2$)')
plt.grid()
plt.show()
````

#### Add legend

````python
t2 = np.linspace(0,1,10*N)
x2 = t2**2
plt.errorbar(t,x,xerr=dt,yerr=dx,fmt='o',capsize=4,label='Data points with uncertainties')
plt.plot(t2,x2,'k--',label='Regression curve')
plt.legend()
plt.xlabel('Time $t$ (s)')
plt.ylabel('Acceleration $a$ (m/s$^2$)')
plt.grid()
plt.show()
````

#### Add a title and make a better resolution image

````python
plt.figure(dpi=150)
plt.title('Data and regression curve from experiment')
t2 = np.linspace(0,1,10*N)
x2 = t2**2
plt.errorbar(t,x,xerr=dt,yerr=dx,fmt='o',capsize=4,label='Data points with uncertainties')
plt.plot(t2,x2,'k--',label='Regression curve')
plt.legend()
plt.xlabel('Time $t$ (s)')
plt.ylabel('Acceleration $a$ (m/s$^2$)')
plt.grid()
plt.show()
````

#### Examples of bad graphs

````python
import numpy as np
import matplotlib.pyplot as plt
N = 100
t = np.linspace(0,20,N)
x = np.sin(t)
y = 1000+np.sin(t)
plt.figure()
plt.plot(t,x,'b-')
plt.plot(t,y,'r-')
plt.title('Graph hides the nature of time evolution (oscillations)')
plt.show()
````
