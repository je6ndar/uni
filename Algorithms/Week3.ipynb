### b)

```python
def newton(f, df, x0, nmax):
    x = x0
    iterates = [x0]
    for n in range(nmax):
        x = x - f(x)/df(x)
        iterates.append(x)
    return iterates
```

```python
f = lambda x: -0.02440529 + 0.39603524*x
df = lambda x: 0.39603524
X = newton(f, df, 0, 5)
X[-1]
```

```python
def secant(f, x0, x1, n_max):
    a = x0
    b = x1
    f_a = f(a)
    f_b = f(b)
    X = [a, b]
    for n in range(1,n_max, 1):
        d = f_b * (b-a)/(f_b - f_a)
        a = b
        f_a = f_b
        b = b - d
        f_b = f(b)
        X.append(b)
    return X
```
